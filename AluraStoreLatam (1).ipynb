```python
import matplotlib.pyplot as plt

# Crear el gráfico de barras
plt.figure(figsize=(8, 5))
barras = plt.bar(tabla_promedios["Tienda"], tabla_promedios["Promedio de Calificación"], color='skyblue')

# Agregar título y etiquetas
plt.title("Promedio de Calificación por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Promedio de Calificación")
plt.ylim(0, 5)  # Escala de calificación de 0 a 5

# Mostrar los valores encima de cada barra (con 2 decimales)
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width() / 2, altura + 0.05,
             "{:.2f}".format(altura), ha='center', va='bottom')


# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calcular los promedios de los costos de envío para cada tienda
envio_tienda = tienda["Costo de envío"].mean()
envio_tienda2 = tienda2["Costo de envío"].mean()
envio_tienda3 = tienda3["Costo de envío"].mean()
envio_tienda4 = tienda4["Costo de envío"].mean()

# Calcular los totales de ventas para cada tienda
dfA = tienda.Precio.sum()
dfB = tienda2.Precio.sum()
dfC = tienda3.Precio.sum()
dfD = tienda4.Precio.sum()

# Crear un DataFrame con los resultados de costo de envío
resultados_envio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [envio_tienda, envio_tienda2, envio_tienda3, envio_tienda4]
})

# Crear un DataFrame con los resultados de ventas
resultados_ventas = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ventas': [dfA, dfB, dfC, dfD]
})

# Configuración del gráfico
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico de barras para el Costo de Envío (columna)
sns.barplot(x='Tienda', y='Costo de Envío Promedio', data=resultados_envio, palette='viridis', ax=ax1, alpha=0.6)
ax1.set_xlabel("Tienda")
ax1.set_ylabel("Costo de Envío Promedio", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Crear el segundo eje Y para las ventas (línea)
ax2 = ax1.twinx()
ax2.plot(resultados_ventas['Tienda'], resultados_ventas['Ventas'], color='red', marker='o', linestyle='-', label="Ventas", linewidth=2)
ax2.set_ylabel("Ventas", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Títulos y etiquetas
plt.title("Comparación de Costo de Envío Promedio (Columna) y Ventas (Línea) entre Tiendas")

# Mostrar el gráfico
plt.show()

```

Este proyecto tiene como objetivo evaluar a partir de los datos recolectados de 4 tiendas cuál es la tienda que  estrategicamente debe ser vendida. La decision evalua los ingresos y ventas promedio que genera cada tienda, los articulos y categorias de items mas vendidos, las calificaciones promedios otorgadas por los clientes y los costos de envios.

Iniciamos el análisis destacando las ventas promedio que genera la Tienda 1 y 2 en comparación de las tiendas 3 y 4, estos resultados indican que existe una alta concurrencia para esas tiendas.

En la siguiente tabla comparamos los ingresos por ventas que antes discutimos con los costos de envio, el gráfico deja en evidencia que hay una relación inversa entre ambos valores, es decir, la tienda con más ingresos por ventas  es la que más costo de envio genera ( Tienda 1), y la que cuenta con menos ingresos por ventas tiene el costo de envio menor ( Tienda 4). Evaluando el Histograma resulta más favorecedor para un cliente comprar en la tienda 4 y 3 ya que gastara menos en el costo de envio. Entre las tiendas 2 y 3 no hay una diferencia significativa con los costos de envio.

Analizando los articulos más vendidos en las 4 tiendas se puede observar que las ventas se concentran en las categorias de Electrónicos, Muebles, Electrodomésticos y Juguetes. Las primeras 3 categorias concentran los artículos de precios más elevados que las tiendas pueden ofertar. La tienda 1 y 3 tienen 51% de las ventas en esas categorias mientras que la tienda 2 tiene 49.50 % y tienda 4 50.30%. El ingreso en ventas de la tienda 2 sobre la tienda 3 y 4 se explica por una alta preferencia de los usuarios en comprar sus instrumentos músicales en la tienda 2, los instrumentos músicales se encuentran dentro de los 15 productos con precios más elevados que se pueden adquirir en las tiendas. La tienda 4 concentra el porcentaje más alto en las categoría ¨juguetes" y "articulos para el hogar" que contiene los articulos con precios más economicos ofertados en las 4 tiendas.
